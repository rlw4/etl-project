```python
import pandas as pd
```

```python
csv_path = "mlb_data.csv"

# Import the books.csv file as a DataFrame
mlb_df = pd.read_csv(csv_path, encoding="utf-8")
mlb_df.head()
```

```python
mlb_atl = mlb_df[mlb_df['home_team']== 'atl']
mlb_atl
```

```python
cleaned_mlb_atl = mlb_atl.loc[:,["home_team","away_team","home_team_score", "away_team_score","date", "year", "season"]]
cleaned_mlb_atl
```

```python
sorted_mlb_atl = cleaned_mlb_atl.sort_values("year", ascending=True)
sorted_mlb_atl.head()
```
