```python
#add time 
mlb_games_df2=mlb_games_df1.loc[(mlb_games_df1["date"]=='""Mon')].replace('""Mon', 'Mon 7pm')
mlb_games_df2
```

```python
#add time 
mlb_games_df4=mlb_games_df1.loc[(mlb_games_df1["date"]=='""Wed')].replace('""Wed', 'Wed 7pm')
mlb_games_df4
```

```python
#add time 
mlb_games_df5=mlb_games_df1.loc[(mlb_games_df1["date"]=='""Thu')].replace('""Thu', 'Thu 7pm')
mlb_games_df5
```

```python
#add time 
mlb_games_df6=mlb_games_df1.loc[(mlb_games_df1["date"]=='""Fri')].replace('""Fri', 'Fri 7pm')
mlb_games_df6
```

```python
#add time 
mlb_games_df7=mlb_games_df1.loc[(mlb_games_df1["date"]=='""Sat')].replace('""Sat', 'Sat 1pm')
mlb_games_df7
```

```python
#add time 
mlb_games_df8=mlb_games_df1.loc[(mlb_games_df1["date"]=='""Sun')].replace('""Sun', 'Sun 1pm')
mlb_games_df8
```

```python
#combine all weekdays data
df=mlb_games_df8.append(mlb_games_df7)
df1=df.append(mlb_games_df6)
df2=df1.append(mlb_games_df5)
df3=df2.append(mlb_games_df4)
df4=df3.append(mlb_games_df3)
df5=df4.append(mlb_games_df2)
df5
```

```python
#remove Preseason data
df6=df5.loc[df5["season"]!="Preseason"]
df6
```

```python
#combine time data to one column
df6["time"]=df6["date"]+df6["game_id"]+df6['year"']
df6
```

```python
#rename columns names
df6.columns=["away_team","away_team_score","day","date","game_id","home_team","home_team_score","season","year","time"]
df6
```

```python
#get game_id data
df7=pd.DataFrame(df6["game_id"])
df7
```

```python
#remove all quotes
df8=df6.loc[:,df6.dtypes==object].apply(lambda s:s.str.replace('"', " "))
df8
```

```python
#add back game_id data
df8["game_id"]=df7
df8
```

```python
#convert time to timestamp
from datetime import date,time, datetime as dt
timestamps=[]
for n in range(0,908):
    date = df8['time'][n][:-1]
    datetime_obj=dt.strptime(date,'%a %I%p %b %d %Y')
    timestamp=(datetime_obj-dt(1970,1,1)).total_seconds()
    timestamps.append(timestamp)
    n+=1
print(timestamps)
```

```python
#add timestamp column
df8["timestamp"]=timestamps
df8
```

```python
path =  "Scripts/Resources/atlanta_weather_history.json"
openweather_data = json.load(open(path))
gameweather_data = pd.DataFrame()

for timestamp in df8["timestamp"]:
    
    
    
    for value in openweather_data:
        try:
            if timestamp == value['dt']:
                weather = [[value['dt'],
                            value['main']['temp'], 
                            value['main']['humidity'], 
                            value['clouds']['all'],
                            value['weather'][0]['main'],
                            value['wind']['speed']]]
                
                
                gameweather_data = gameweather_data.append(weather)
            
    
        except (KeyError, IndexError):
            print("Missing field/result... skipping.")

#end for
gameweather_data.columns = ['Timestamp','Temperature', 'Humidity', 'Cloudiness', 'Weather', 'Wind Speed']


display(gameweather_data)
```

```python
gameweather_data.reset_index(drop=True)
gameweather_data
```

```python
timestamp=gameweather_data["Timestamp"].tolist()
temperature=gameweather_data["Temperature"].tolist()
humidity=gameweather_data["Humidity"].tolist()
cloudiness=gameweather_data["Cloudiness"].tolist()
weather=gameweather_data["Weather"].tolist()
wind_speed=gameweather_data["Wind Speed"].tolist()
home_team_score=df8["home_team_score"].tolist()
home_team=df8["home_team"].tolist()
away_team=df8["away_team"].tolist()
away_team_score=df8["away_team_score"].tolist()
season=df8["season"].tolist()
time=df8["time"].tolist()
game_id = np.array(df8["game_id"], dtype=np.int)

atl_baseball_games=pd.DataFrame({"Away_Team":away_team,"Away_Team_Score":away_team_score,"Home_Team":home_team,"Home_Team_Score":home_team_score,"Game_ID":game_id,"Time":time,"Timestamp":timestamp,"Temperature":temperature,"Humidity":humidity,"Cloudiness":cloudiness,"Weather":weather,"Wind Speed":wind_speed})
atl_baseball_games
```
